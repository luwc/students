```python
!pip install yfinance
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from keras import Sequential
from keras.layers import SimpleRNN, Dropout, Dense
np.random.seed(42)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import tensorflow as tf
media = yf.Ticker("2454.TW")
hist = media.history(period="10Y", interval="1d", start=None, end=None, actions=True, auto_adjust=True, back_adjust=False)
hist['Close'].plot()
hist.head()
```

```python
hist_len=len(hist)
print("data observation", hist_len)
N=250 #想要測試筆數
x_train_set=hist.iloc[0:hist_len-N,3].values
print(x_train_set)
x_test_set=hist.iloc[hist_len-N:hist_len,3].values
print(x_test_set)

#加工資料
sc=MinMaxScaler(feature_range=(0,1))
x_train_set=sc.fit_transform(x_train_set.reshape(-1,1))
x_test_set=sc.transform(x_test_set.reshape(-1,1))
print(x_train_set)
print(x_test_set)
```

```python
look_back = 60
x_train=[]
y_train=[]
x_test=[]
y_test=[]
for i in range(look_back, len(x_train_set)):
    x_train.append(x_train_set[i-look_back:i,0])
    y_train.append(x_train_set[i,0])
np.random.seed(7)
np.random.shuffle(x_train)
np.random.seed(7)
np.random.shuffle(y_train)
x_train=np.array(x_train)
y_train=np.array(y_train)
x_train=np.reshape(x_train,(x_train.shape[0],look_back,1))
# test set
x_test_set = x_test_set.reshape(-1, 1)
for i in range(look_back, len(x_test_set)):
    x_test.append(x_test_set[i-look_back:i, 0])
    y_test.append(x_test_set[i, 0])

x_test=np.array(x_test)
y_test=np.array(y_test)
x_test=np.reshape(x_test,(x_test.shape[0],look_back,1))



```

```python
#SimpleRNN model##
model=tf.keras.Sequential([
    SimpleRNN(units=80, return_sequences=True),
    Dropout(0.2),
    SimpleRNN(100),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# Train the model and store the training history
history = model.fit(x_train, y_train, epochs=100, batch_size=32) # Example training parameters

loss=history.history["loss"]
plt.plot(loss, label="Training Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.show()


predicted_price=model.predict(x_test)
predicted_price=sc.inverse_transform(predicted_price)
print(predicted_price)
real_price=sc.inverse_transform(x_test_set[look_back:])
plt.plot(real_price, color="red", label="Real Price")
plt.plot(predicted_price, color="blue", label="Predicted Price")
plt.title("Stock Price Prediction")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
#SimpleRNN model##
model=tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(units=80, return_sequences=True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.SimpleRNN(100),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
# Train the model and store the training history
history = model.fit(x_train, y_train, epochs=100, batch_size=32)  # Example training parameters

loss=history.history["loss"]
plt.plot(loss, label="Training Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.show()


predicted_price=model.predict(x_test)
predicted_price=sc.inverse_transform(predicted_price)
print(predicted_price)
real_price=sc.inverse_transform(x_test_set[look_back:])
plt.plot(real_price, color="red", label="Real Price")
plt.plot(predicted_price, color="blue", label="Predicted Price")
plt.title("Stock Price Prediction")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```

```python
#模型評模型評價
from sklearn.metrics import mean_squared_error, mean_absolute_error # Import necessary functions

mse=mean_squared_error(real_price,predicted_price)
print('MSE(均方誤差)=', mse)
rmse=np.sqrt(mse)
print('MSE(均方根誤差)=', rmse)
mae=mean_absolute_error(real_price,predicted_price)
print('MSE(平均絕對誤差)=', mae)
```
